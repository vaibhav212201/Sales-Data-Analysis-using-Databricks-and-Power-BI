Sales Data Analysis using Databricks & Pandas


```python
# Import Libraries
import pandas as pd
# Load Dataset
df = pd.read_csv("/Volumes/workspace/default/data/sales_data_sample.csv",
                 encoding='latin1')
```

```python
# Data Understanding
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns)
```

```python
# Data Cleaning

# Convert date column
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Drop unnecessary column
df.drop(columns='ADDRESSLINE2', inplace=True)

# Check Null Values
print("\nNull Values:")
print(df.isnull().sum())

# Check Duplicates
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
# KPI Analysis
total_revenue = df['SALES'].sum()
total_orders = df['ORDERNUMBER'].nunique()
units_sold = df['QUANTITYORDERED'].sum()

print("\nTotal Revenue:", round(total_revenue, 2))
print("Total Orders:", total_orders)
print("Units Sold:", units_sold)
```

```python
# Product Category Analysis
print("\nRevenue by Product Category:")
print(df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False))
```

```python
# Country Analysis
print("\nRevenue by Country:")
print(df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10))
```

```python
# Territory Analysis
print("\nRevenue by Territory:")
print(df.groupby('TERRITORY')['SALES'].sum().sort_values(ascending=False))
```

```python
# Product Analysis
print("\nTop 10 Products:")
print(
    df.groupby('PRODUCTCODE')['SALES']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print("\nBottom 10 Products:")
print(
    df.groupby('PRODUCTCODE')['SALES']
      .sum()
      .sort_values()
      .head(10)
)
```

```python
# Monthly Analysis
print("\nRevenue by Month:")
print(df.groupby('MONTH_ID')['SALES'].sum())
```
